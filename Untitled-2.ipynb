```python
from rdkit import Chem

def detect_functional_groups(smiles):
    """
    Detect functional groups in a molecule from SMILES.
    
    Returns a dictionary of {group_name: count}
    """
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        raise ValueError(f"Invalid SMILES: {smiles}")
    
    # Define SMARTS patterns for each functional group
    patterns = {
        # Hydrocarbons
        'Alkane': '[CX4]',  # sp3 carbon with 4 bonds
        'Alkene': '[CX3]=[CX3]',  # C=C double bond
        'Alkyne': '[CX2]#[CX2]',  # C≡C triple bond
        'Arene': 'c',  # Aromatic carbon
        
        # Halides
        'Chloroalkane': '[CX4][Cl]',  # C-Cl (aliphatic)
        'Fluoroalkane': '[CX4][F]',  # C-F (aliphatic)
        'Bromoalkane': '[CX4][Br]',  # C-Br (aliphatic)
        'Iodoalkane': '[CX4][I]',  # C-I (aliphatic)
        
        # Oxygen-containing groups
        'Acyl chloride': '[CX3](=[OX1])[Cl]',  # R-C(=O)-Cl
        'Alcohol': '[CX4][OX2H]',  # R-OH (aliphatic)
        'Aldehyde': '[CX3H1](=O)[#6]',  # R-CHO
        'Ketone': '[#6][CX3](=O)[#6]',  # R-C(=O)-R'
        'Carboxylic acid': '[CX3](=O)[OX2H1]',  # R-COOH
        'Acid anhydride': '[CX3](=[OX1])[OX2][CX3](=[OX1])',  # R-C(=O)-O-C(=O)-R'
        'Ester': '[#6][CX3](=O)[OX2][#6]',  # R-C(=O)-O-R'
        'Ether': '[OD2]([#6])[#6]',  # R-O-R' (not in C=O)
        'Phenol': '[OX2H][c]',  # Ar-OH (aromatic)
        'Enol': '[OX2H][CX3]=[CX3]',  # C=C-OH
        
        # Nitrogen-containing groups
        'Amine': '[NX3;H2,H1;!$(NC=O)]',  # R-NH2 or R2-NH (not amide)
        'Amide': '[NX3][CX3](=[OX1])',  # R-C(=O)-N
        'Nitrile': '[NX1]#[CX2]',  # R-C≡N
        'Imide': '[CX3](=[OX1])[NX3][CX3](=[OX1])',  # O=C-NH-C=O
        'Imine': '[CX3]=[NX2;!$(N=O)]',  # R-C=N-R' (not nitro)
        
        # Sulfur-containing groups
        'Thiol': '[#16X2H]',  # R-SH
        'Sulfide': '[#16X2]([#6])[#6]',  # R-S-R'
    }
    
    results = {}
    for group_name, smarts in patterns.items():
        pattern = Chem.MolFromSmarts(smarts)
        matches = mol.GetSubstructMatches(pattern)
        results[group_name] = len(matches)
    
    return results


def has_functional_group(smiles, group_name):
    """Check if a specific functional group is present."""
    results = detect_functional_groups(smiles)
    return results.get(group_name, 0) > 0


# Test examples
if __name__ == "__main__":
    test_molecules = {
        "Butane": "CCCC",
        "2-Butene": "CC=CC",
        "1-Butyne": "C#CCC",
        "Benzene": "c1ccccc1",
        "Chloroethane": "CCCl",
        "Acetyl chloride": "CC(=O)Cl",
        "Ethanol": "CCO",
        "Acetaldehyde": "CC=O",
        "Acetone": "CC(=O)C",
        "Acetic acid": "CC(=O)O",
        "Acetic anhydride": "CC(=O)OC(=O)C",
        "Ethyl acetate": "CC(=O)OCC",
        "Diethyl ether": "CCOCC",
        "Phenol": "Oc1ccccc1",
        "Methylamine": "CN",
        "Acetamide": "CC(=O)N",
        "Acetonitrile": "CC#N",
        "Succinimide": "O=C1CCC(=O)N1",
        "Ethanethiol": "CCS",
        "Dimethyl sulfide": "CSC",
        "Aspirin": "CC(=O)Oc1ccccc1C(=O)O",
    }
    
    print("Functional Group Detection")
    print("=" * 80)
    
    for name, smiles in test_molecules.items():
        print(f"\n{name}: {smiles}")
        groups = detect_functional_groups(smiles)
        # Only show groups that are present
        present = {k: v for k, v in groups.items() if v > 0}
        for group, count in present.items():
            print(f"  ✓ {group}: {count}")
```
